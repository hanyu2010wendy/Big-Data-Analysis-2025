```python
import pandas as pd
import numpy as np

df = pd.read_csv(DATA_DIR / "wdi" / "WDI_course_subset.csv")

#### Use Fiji as the example country.
df_FJ = df[df["Country Name"] == "Fiji"]

WDI_FJ = df_FJ.drop(columns="Indicator Code").melt(
    id_vars=["Country Name", "Country Code", "Indicator Name"],
    var_name="Year"
).pivot_table(
    values="value",
    index=["Country Name", "Country Code", "Year"],
    columns="Indicator Name",
).reset_index().rename_axis("", axis=1)

### Convert Year from string-like values to integers.
WDI_FJ["Year"] = WDI_FJ["Year"].astype(str).astype(int)
WDI_FJ.head()
```

```python
isna_data = WDI_FJ.isna().sum().sort_values(ascending=True)

isna_data.to_csv(MODULE_OUTPUT_DIR / "isna_data_FJ.csv", index=True)
isna_data
```

# 1. Based on the data from Homework1, select appropriate variables to create a scatter plot/bubble chart and a bar chart.

```python
fig_FJ = WDI_FJ[WDI_FJ["Year"] > 2010]
fig_FJ.head()
```

```python
###import matplotlib
import matplotlib.pyplot as plt

###use plt.scatter

###c stands for color. alpha means the transparency.
plt.scatter(fig_FJ['Year'],
            fig_FJ['Merchandise trade (% of GDP)'],
            c= "orange",
            alpha=0.5
)
```

# 2. Based on the data in Homework1, select two appropriate variables to make a line chart and export it.

```python
### Plot the first line.
plt.plot(
    fig_FJ["Year"],
    fig_FJ["Merchandise trade (% of GDP)"],
    linestyle="-.",
    linewidth=1,
    marker=".",
    color="#0f0f0f80",
)

### Plot the second line.
plt.plot(
    fig_FJ["Year"],
    fig_FJ["Manufactures exports (% of merchandise exports)"],
    color="yellow",
)

### Add labels and a title.
plt.xlabel("Year")
plt.ylabel("Unit: %")
plt.title("Fiji: Two Trade Indicators")

### Set x-axis ticks and add a legend.
plt.xticks(np.arange(2010, 2022, step=5))
plt.legend(
    [
        "Merchandise trade (% of GDP)",
        "Manufactures exports (% of merchandise exports)",
    ],
    loc="center left",
)

plt.savefig(MODULE_OUTPUT_DIR / "Homework3_line_chart.png", bbox_inches="tight")
```
