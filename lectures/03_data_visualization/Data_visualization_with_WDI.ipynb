This repository includes a lightweight WDI subset for teaching purposes.

# **Part I：Data Preprocessing**

```python
import pandas as pd

df = pd.read_csv(DATA_DIR / "wdi" / "WDI_course_subset.csv")

# Use the United States as the example country.
df_US = df[df["Country Name"] == "United States"]

WDI_US = df_US.drop(columns="Indicator Code").melt(
    id_vars=["Country Name", "Country Code", "Indicator Name"],
    var_name="Year"
).pivot_table(
    values="value",
    index=["Country Name", "Country Code", "Year"],
    columns="Indicator Name",
).reset_index().rename_axis("", axis=1)

WDI_US["Year"] = WDI_US["Year"].astype(str).astype(int)
WDI_US.head()
```

```python
WDI_US.dtypes
```

```python
isna_data = WDI_US.isna().sum().sort_values(ascending=True)
isna_data.to_csv(MODULE_OUTPUT_DIR / "isna_data_US.csv", index=True)
isna_data
```

# Q1 How to create a scatter plot?

```python
### Import the matplotlib library for plotting. numpy is used for simple calculations, and it will be imported here as well.
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Example
### The data in this example consists of random numbers and is for demonstration purposes only.

### Fixing random state for reproducibility
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y, alpha=0.5)
```

# Try drawing it with your own data!

```python
### Define the data that can be used to draw a graph. Example: data starts from 1990.
fig_US = WDI_US[WDI_US["Year"] > 1990]

fig_US.head()
```

```python
### Import matplotlib
import matplotlib.pyplot as plt

### Using plt.scatter, pass in the x and y data to draw the plot

### c means color. alpha is used to set transparency.

plt.scatter(fig_US['Year'],
            fig_US['Merchandise trade (% of GDP)'],
            c= "orange",
            alpha=0.5
)
```

# Q2 How to create a line plot?

```python
### Use plt.plot

plt.plot(fig_US['Year'], # x
        fig_US['Merchandise trade (% of GDP)'], # y
        linestyle = '--',
        linewidth="1", 
        marker = "1", # Used to set the markers on the line
        color="plum"
)


plt.xlabel("Year")

plt.ylabel("Merchandise trade (% of GDP)")

plt.title("1960-2022 US Merchandise trade (% of GDP)")

plt.grid(visible=True, which='major', axis='both')

plt.tick_params(direction='in',
                length=4, 
                width=2, 
                colors='black',
                grid_color='black', 
                grid_alpha=0.1
)

plt.xticks(range(1990, 2022, 5))

plt.savefig(MODULE_OUTPUT_DIR / "lineplot.png", bbox_inches="tight")
```

```python
### Plot two lines on the same plot.

### Plot the first line
plt.plot(fig_US['Year'],
        fig_US['Merchandise trade (% of GDP)'],
        #linestyle = '-.',
        #linewidth="1",
        #marker = ".",
        color="#0f0f0f80"
)
### Plot the second line        
plt.plot(fig_US['Year'],
        fig_US['Manufactures exports (% of merchandise exports)'],
        color="Yellow"
)

### Add X and Y-axis labels，and add titles
#plt.xlabel("Year")
#plt.ylabel("Unit: %")
#plt.title("")

### Set ticks.
# plt.xticks(np.arange(1960, 2022, step=5))
### Add a legend.
plt.legend(['Merchandise trade (% of GDP)', 'Manufactures exports (% of merchandise exports)'], loc="right")
```
