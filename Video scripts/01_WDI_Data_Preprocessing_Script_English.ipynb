# Episode 1: Build a WDI Panel Dataset in Python

## Video purpose
Turn the lecture notebook into a recording-friendly English teaching script.

- Source notebook: `archive/legacy_wdi_materials/WDI data analysis.ipynb`
- Estimated video length: 10-14 minutes
- Core promise: viewers learn how to go from raw WDI data to a clean country-level panel dataset.


## Suggested YouTube title
**Build a WDI Panel Dataset in Python | Big Data Analysis with Python**

## Opening hook
"In this video, I will show you how to turn a raw World Bank dataset into a clean panel dataset that is ready for analysis, visualization, and future research."


## Segment 1: Explain the dataset and the goal

### What to say
Start by explaining that the raw WDI file is informative but not analysis-ready. The purpose of the lesson is to build a reusable panel dataset, not only to read a CSV file.

### What to show on screen
- the raw file structure
- the final panel structure
- a short explanation of why panel data matters


## Segment 2: Read the data and filter one country

### What to say
Tell viewers that using one country as a worked example keeps the logic simple. The same workflow can later be repeated for any other country.

### What to show on screen
- `read_csv`
- `df.head()`
- filter the United States


## Segment 3: Reshape wide data into panel data

### What to say
Explain that `melt` turns year columns into rows, while `pivot_table` turns indicator names into columns. This is the key transformation in the lesson.

### What to show on screen
- the `melt` step
- the `pivot_table` step
- converting `Year` to integer


## Segment 4: Export and inspect missing values

### What to say
Explain that exporting the cleaned dataset is part of reproducible workflow. Then emphasize that checking missing values is essential before visualization or regression.

### What to show on screen
- export `WDI_US.csv`
- count missing values
- mention the bonus exercise


## Closing script
"In this lesson, we read WDI data, filtered one country, reshaped the dataset into panel form, and checked missing values. This is a small workflow, but it is the foundation for many later data-analysis tasks."
